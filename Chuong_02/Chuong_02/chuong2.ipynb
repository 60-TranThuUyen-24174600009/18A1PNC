```python
import xml.etree.ElementTree as ET

root = ET.Element("Students")

sv1 = ET.SubElement(root, "student", id="SV001")
ET.SubElement(sv1, "name").text = "Nguyen Van A"
ET.SubElement(sv1, "year").text  = "2002"
ET.SubElement(sv1, "class").text = "CNTT1"
ET.SubElement(sv1, "gender").text = "Nam"

sv2 = ET.SubElement(root, "student", id="SV002")
ET.SubElement(sv2, "name").text = "Tran Thi B"
ET.SubElement(sv2, "year").text = "2003"
ET.SubElement(sv2, "class").text = "CNTT2"
ET.SubElement(sv2, "gender").text = "Nu"

tree = ET.ElementTree(root)
tree.write("students.xml", encoding="utf-8", xml_declaration=True)

print("Đã tạo file students.xml thành công!")
```

```python
from xml.dom import minidom
import os

path = os.path.join(os.getcwd(), "sample.xml")
print("Đường dẫn tới file:", path)

doc = minidom.parse(path)
root = doc.documentElement
print("Tên thẻ gốc:", root.tagName)

company_name = root.getElementsByTagName("name")[0].firstChild.data
print("Tên công ty:", company_name)

staffs = root.getElementsByTagName("staff")
print("\nDanh sách nhân viên:")
for staff in staffs:
    staff_id = staff.getAttribute("id")
    name = staff.getElementsByTagName("name")[0].firstChild.data
    salary = staff.getElementsByTagName("salary")[0].firstChild.data
    print(f"- ID: {staff_id}, Tên: {name}, Lương: {salary}")

```

```python
import requests
import csv 
import xml.etree.ElementTree as ET

url = "https://vnexpress.net/rss/tin-moi-nhat.rss"

response = requests.get(url)
response.rncoding = 'utf-8'

root = ET.fromstring(response.text)

with open("rss_data.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "Link"])
                     
    for item in root.findall(".//item"):
        title = item.find("title").text if item.find("title") is not None else ""
        link = item.find("link").text if item.find("link") is not None else ""


        writer.writerow([title, link])

print(" Đã ghi dữ liệu RSS vào file rss_data.csv thành công!")
```
